В большой текстовой строке подсчитать количество встречаемых слов и вернуть 10 самых частых. Не учитывать знаки препинания и регистр символов. За основу возьмите любую статью из википедии или из документации к языку.

```python
import re
from collections import Counter

текст = """
Википе́дия — общедоступная многоязычная универсальная интернет-энциклопедия со свободным контентом[комм. 2], реализованная на принципах вики.

Владелец сайта — американская некоммерческая организация «Фонд Викимедиа», имеющая 37 региональных представительств. Название энциклопедии образовано от английских слов wiki («вики»; в свою очередь заимствовано из гавайского языка, в котором оно имеет значение «быстро») и encyclopedia («энциклопедия»).

Запущенная в январе 2001 года Джимми Уэйлсом и Ларри Сэнгером[6], Википедия является самым крупным и наиболее популярным[7] справочником в Интернете[8][9][10]. По объёму сведений и тематическому охвату Википедия считается самой полной энциклопедией из когда-либо создававшихся за всю историю человечества[11][12]. Одним из основных достоинств Википедии как универсальной энциклопедии является возможность представления информации на родном языке пользователя[13]. В Википедии есть 334 раздела на различных языках мира, а также сотни языковых разделов, находящихся в инкубаторе. Она содержит более 60 миллионов статей[1]. Интернет-сайт Википедии входит в топ-10 самых посещаемых сайтов в мире[14].

Главной особенностью Википедии является то, что создавать и редактировать статьи в ней может любой пользователь Интернета[комм. 3]. Все вносимые такими добровольцами изменения незамедлительно становятся доступными для просмотра всем посетителям сайта. В декабре 2013 года в заявлении ЮНЕСКО по случаю награждения Джимми Уэйлса, основателя Википедии, Золотой медалью Нильса Бора про Википедию было сказано, что она является «символом эпохи взаимодействия, в которую мы живём, и это не просто инструмент, это воплощение мечты, столь же древней, как само человечество, как собрания Александрийской библиотеки»[15][16].

Надёжность и точность Википедии вызывают вопросы[17]. Википедию критикуют за возможность добавления ложной или непроверенной информации[18] и вандализма на её страницах. Однако научные исследования свидетельствуют о том, что в Википедии следы актов вандализма обычно оперативно устраняются[19][20].

Раздел Википедии на русском языке, по состоянию на 25 августа 2023 года, насчитывает 1 934 445 статей различной тематики, занимая 7-е место по количеству статей среди всех языковых разделов.
"""

# Преобразовать текст в нижний регистр и удалить знаки препинания
текст = текст.lower()
текст = re.sub(r'[^\w\s]', '', текст)

# Разделить текст на слова
слова = текст.split()

# Подсчитать количество встречаемых слов
частоты = Counter(слова)

# Получить 10 самых часто встречающихся слов
самые_частые = частоты.most_common(10)

# Вывести результат
for слово, количествоВстреч in самые_частые:
    print(f"{слово}: {количествоВстреч}")
```

Создайте словарь со списком вещей для похода в качестве ключа и их массой в качестве значения. Определите какие вещи влезут в рюкзак передав его максимальную грузоподъёмность. Достаточно вернуть один допустимый вариант. *Верните все возможные варианты комплектации рюкзака.

```python
from itertools import combinations

# Функция для поиска всех возможных вариантов комплектации рюкзака
def варианты_комплектации(вещи, макс_грузоподъемность):
    допустимые_комбинации = []
    
    # Перебираем все возможные комбинации вещей
    for i in range(1, len(вещи) + 1):
        комбинации = combinations(вещи.keys(), i)
        for комбинация in комбинации:
            масса_комбинации = sum(вещи[предмет] for предмет in комбинация)
            if масса_комбинации <= макс_грузоподъемность:
                допустимые_комбинации.append(комбинация)
    
    return допустимые_комбинации

# Пример использования
вещи = {
    "спальник": 1.5,
    "палатка": 3.0,
    "еда": 2.5,
    "вода": 2.0,
    "кастрюля": 0.5
}

макс_грузоподъемность = 6.0

допустимые_комбинации = варианты_комплектации(вещи, макс_грузоподъемность)

# Выводим допустимые комбинации
for комбинация in допустимые_комбинации:
    масса_комбинации = sum(вещи[предмет] for предмет in комбинация)
    print(f"Вещи: {', '.join(комбинация)}")
    print(f"Общая масса: {масса_комбинации} кг")
    print()
```
